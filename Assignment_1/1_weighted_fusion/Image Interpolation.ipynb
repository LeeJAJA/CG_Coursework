```python
import numpy as np
import cv2
from  matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('cat.jpg') 
img_1 = cv2.resize(img,(1024,1024))  
img_1 = cv2.cvtColor(img_1, cv2.COLOR_RGB2BGR)
img = cv2.imread('tiger.jpg') 
img_2 = cv2.resize(img,(1024,1024))  
img_2 = cv2.cvtColor(img_2, cv2.COLOR_RGB2BGR)
plt.imshow(np.hstack([img_1, img_2]))
plt.show()
```

```python
for i in range(11):
    img_fusion = i/10*img_1 + (10-i)/10*img_2
    img_fusion = img_fusion.astype(np.uint8)
    cv2.imwrite(f'./output/fused_img_{i}_{10-i}.png', cv2.cvtColor(img_fusion, cv2.COLOR_BGR2RGB))
    plt.imshow(img_fusion)
    plt.show()
```
